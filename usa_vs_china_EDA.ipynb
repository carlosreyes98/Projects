MILITARY POWER

```python
power1 = army[:4]
power2 = army[4:12]

sns.barplot(y = power2["Description"], x = power2["China"], color = "red")
sns.barplot(y = power2["Description"], x = power2["EEUU"],color="blue")
plt.xlabel("")
plt.ylabel("")
plt.title("Army Comparison",fontsize=15)
plt.show()

sns.barplot(y = power1["Description"], x = power1["EEUU"],color="blue")
sns.barplot(y = power1["Description"], x = power1["China"], color = "red")
plt.xlabel("")
plt.ylabel("")
```

MIGRATION

```python
migration = migration.drop(["Country Name","Indicator Name","Indicator Code"],axis=1).dropna(how="all",axis=1)
usa = migration[251:252]
china = migration[40:41]

years = list(migration.columns)
years.pop(0)
years = list(map(int, years))

usa = usa.values.tolist()[0]
usa.pop(0)
usa = list(map(int, usa))

china = china.values.tolist()[0]
china.pop(0)
china = list(map(int, china))

sns.lineplot(x=years,y=china,color="red")
sns.lineplot(x=years,y=usa,color="blue")
plt.title("Migration History USA vs China(in Millions)",fontsize=11)
plt.text(2010,150000,"China",color="red") and plt.text(2010,1600000,"USA",color="blue")
```
